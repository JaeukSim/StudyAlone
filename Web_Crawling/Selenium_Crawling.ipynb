# 멜론에서 음악차트 크롤링 하기

```python
from selenium import webdriver
driver = webdriver.Chrome('D:\심재욱\심재욱\AIschool\심화과정\실습자료\실습\심화과정실습\Second Weak\chromedriver')
driver.implicitly_wait(3)
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
rank_infos=driver.find_elements_by_css_selector(".service_list_song table tbody tr") # class="service_list_song" 클래스 하위에 있는 속성 table tbody tr을 모두 크롤링 함
data=[] #크롤링한 데이터를 저장할 배열생성
rank=1 # 보기좋게 순서 매기기 위해서 랭킹 순번 생성
for tr in rank_infos:
    title = tr.find_element_by_css_selector('.wrap_song_info .ellipsis.rank01 a').text
    singer= tr.find_element_by_css_selector('.wrap_song_info .ellipsis.rank02 a').text
    print(title,"  ", singer)
    data.append([rank, title, singer])
    rank +=1

with open('from Selenium_Crawling_melon_chart_ranking.csv', 'w', encoding='utf-8') as f:    #파일을 쓰기모드로 생성하기>> 파일이름은 from Selenium_Crawling_melon_chart_ranking.csv
    f.write('순위, 노래제목, 가수')
    for i in data:
        f.write('{0},{1},{2}\n '.format(i[0], i[1], i[2]))

driver.close()




```

# 네이버에서 영화별로 네티즌 평점 받아오기 (10page까지)

```python
from selenium import webdriver
import time
driver = webdriver.Chrome('D:\심재욱\심재욱\AIschool\심화과정\실습자료\실습\심화과정실습\Second Weak\chromedriver')
driver.implicitly_wait(3)

max_page=10
movie_scores = dict() #영화별로 평점을 저장할 딕셔너리 배열 생성

for current_page in range(1, max_page+1):
    url = "https://movie.naver.com/movie/point/af/list.nhn?&page={}".format(current_page)  #네이버영화에서 페이지를 넘길때마다 page=페이지숫자 가 적혀있는 공통점 발견 후 코딩
    driver.get(url)
    time.sleep(0.1)

    movies = driver.find_elements_by_css_selector("table.list_netizen tbody tr td.title")
    for movie in movies:
        title = movie.find_element_by_css_selector("a").text
        score = movie.find_element_by_css_selector(".list_netizen_score em").text
        # print(title, "  ", score)
        

        if title not in movie_scores:
            movie_scores[title]=list()     #위에서 만든 딕셔너리 형태의 배열 movie_scores에 각 제목을 키값으로 그리고 해당 value를 배열[]로 생성 (일단 배열은 빈배열)
            # print(movie_scores)
        
        movie_scores[title].append(float(score)) #생성된 딕셔너리 배열의 value란에 위에서 받아온 score를 차곡차곡 추가시킴
        # print(movie_scores.items())
movie_scores_mean = list()
for title, score in movie_scores.items():
    movie_scores_mean.append((title, sum(score)/len(score)))
        
for m in movie_scores_mean:
    print(m[0],"  ",m[1])

driver.close()
```

# 구글 이미지 크롤링


```python
from selenium.webdriver.common.keys import Keys  #Keys 의 k 대문자 조심!!!
import time
import urllib.request
import os

keyword = input("Keyword : ")
total_count = input("How many images do you wanna download : ")

driver = webdriver.Chrome('D:\심재욱\심재욱\AIschool\심화과정\실습자료\실습\심화과정실습\Second Weak\chromedriver')
driver.get("http://www.google.co.kr/imghp?hl=ko")
driver.implicitly_wait(3)

input_element = driver.find_element_by_name("q")
input_element.send_keys(keyword)
input_element.send_keys(Keys.RETURN)

images = driver.find_elements_by_css_selector('.rg_i.Q4LuWd') #모든 사진이 공통적으로 특수하게 가지고 있는 클래스

count=0

for image in images:
    # image.click() #해당 이미지를 클릭
    time.sleep(0.5)

    count +=1 

    imgURL = image.get_attribute('src') # 이미지의 주소를 src속성으로 받아와서 저장함
    


    if not os.path.exists('google_imgs'): #경로에 있나 확인
        os.makedirs('google_imgs') #없으면 새로운 파일 생성
    
    urllib.request.urlretrieve(imgURL, "google_imgs/"+keyword+str(count)+".jpg")  #해당이미지를 다운로드하는 함수가 urllib.request.urlretrieve(이미지주소, 저장할이름)

    if count == int(total_count):
        break

   


driver.close()
```
