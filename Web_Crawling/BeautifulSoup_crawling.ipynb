```python
import bs4 #beautifulsoup import하기
import urllib.request #Beautifulsoup은 통신기능이 없어서 통신 기능 추가
url = "https://movie.naver.com/"
html = urllib.request.urlopen(url)
# print(html)
bsObj =bs4.BeautifulSoup(html, 'html.parser')
# print(bsObj)

realtime_hotkeyword = bsObj.find_all("div", {"id":"content"})
print(realtime_hotkeyword)
```
