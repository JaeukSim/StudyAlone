```python
import numpy as np 
import matplotlib.pylab as plt


# [-10, 10]
t = np.linspace(-10,10, 100)

# 시그모이드 공식
# Numpy np.exp() 함수를 이용한다. 이 함수는 밑수를 자연상수 e인 지수함수로 변환해줍니다

sig = 1/(1+np.exp(-t))

# t 와 시그모이드 결과값을 Plotting
plt.figure(figsize=(9,3))
plt.plot([-10,10], [0,0], "k-")
plt.plot([-10,10], [0.5, 0.5], "k:") # y축 기준 0.5 실선 생성
plt.plot([-10, 10], [1,1], "k:") # y 축 기준으로 1.0 실선 생성
plt.plot([0,0], [-1.1,1.1], "k-") # Center 기준 0.0 선 생성
plt.plot(t, sig, 'r-', linewidth = 2, label=r"$\sigma(t)= \frac{1}{1+e^{-t}}$")
plt.xlabel('t') # x축 이름
plt.legend(loc = "upper left", fontsize=20)
plt.axis([-10,10, -0.1, 1.1]) # 그래픽 간격
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
x= np.arange(-5.0, 5.0,0.1)
y_1 =sigmoid(0.5*x)
y_2 =sigmoid(x)
y_3 =sigmoid(2*x)
# print("\n y_1 sigmoid", y_1)
# print("\n y_2 sigmoid", y_2)
# print("\n y_3 sigmoid", y_3)

plt.plot(x, y_1, 'r', linestyle='--') # w의 값이 0.5
plt.plot(x, y_2, 'g') # w의 값이 1 일때
plt.plot(x, y_3, 'b', linestyle='--') # w의 값이 2 일때
plt.plot([0,0], [1.0,0.0]) # 가운데 실선 추가
plt.title("Sigmoid Function")
plt.show()
```
